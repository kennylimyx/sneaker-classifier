## Import Libraries

```python
import numpy as np
import pandas as pd
from PIL import Image

import matplotlib.pyplot as plt
import matplotlib.image as img
import seaborn as sns

from sklearn.model_selection import train_test_split
from keras.utils import np_utils
```

## Import Data

```python
rep = pd.read_csv("./datasets/rep_url_1k.csv")
auth = pd.read_csv("./datasets/auth_url_1k.csv")
```

```python
df = pd.concat([rep,auth],axis=0).sample(frac=1)
print(df.shape)
```

```python
#reset index
print(df.loc[0,'url'])
df.reset_index(inplace=True,drop=True)  #drop=True prevents new index col
```

```python
df['name'] = df['url'].map(lambda x: x.split('/')[-1])
```

```python
#check changes
df.head()
```

```python
sns.countplot(df['is_rep']);
```

## Resize Images

```python
#resize images
def resize_img(name, label, resize_config = (150,150)):
    temp_path = f"./datasets/images/{label}/{name}"
    image = Image.open(temp_path)
    
    new_img = image.convert(mode='RGB')                    #ensure all images are RGB
    new_img = new_img.resize(resize_config, Image.LANCZOS) #resizing
    
    #return img as array
    return np.array(new_img)
```

```python
img_array = []
labels = []

for i in range(df.shape[0]):
    print(f"resizing image {i+1} of {df.shape[0]}...")
    
    try:
        img_array.append(resize_img(df.loc[i,'name'], df.loc[i,'is_rep']))
        labels.append(1 if df.loc[i,'is_rep'] == "rep" else 0)
    except FileNotFoundError:
        pass
```

```python
#let's take a look at our resized images
plt.figure(figsize=(20,10))
col=5
for i in range(5):
    plt.subplot(5/col+1,col,i+1)
    plt.imshow(img_array[i])
```

```python
#compare this to the df.head() above
labels[:5]
```

```python
#convert list to np array
X = np.array(img_array)
y = np.array(labels)

#make sure data are all float to prevent any errors when we feed it into machine
X = X.astype('float32')
#scale the data
X /= 255
```
